```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(dr[['Age','Annual Income (k$)']])
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1, 11), wcss, lw = 3)
plt.title('The Elbow Method', fontsize = 25)
plt.xlabel('Number of clusters', fontsize = 15)
plt.ylabel('WCSS', fontsize =15 )
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_predicted = kmeans.fit_predict(dr[['Age','Annual Income (k$)']])
```

```python
import sklearn
sklearn.metrics.calinski_harabasz_score(dr[['Age','Annual Income (k$)']], y_predicted)
```

```python
plt.figure(figsize = (12,8))
dr1=dr[dr.cluster==0]
dr2=dr[dr.cluster==1]
dr3=dr[dr.cluster==2]
dr4=dr[dr.cluster==3]
dr5=dr[dr.cluster==4]
plt.scatter(dr1.Age,dr1['Annual Income (k$)'],color='green', label = 'Cluster 1')
plt.scatter(dr2.Age,dr2['Annual Income (k$)'],color='red', label = 'Cluster 2')
plt.scatter(dr3.Age,dr3['Annual Income (k$)'],color='blue', label = 'Cluster 3')
plt.scatter(dr4.Age,dr4['Annual Income (k$)'],color='yellow', label = 'Cluster 4')
plt.scatter(dr5.Age,dr5['Annual Income (k$)'],color='white', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Annual Income (k$)',fontsize = 15)
plt.title('Clusters', fontsize = 25)
plt.legend(fontsize = 15)
```
